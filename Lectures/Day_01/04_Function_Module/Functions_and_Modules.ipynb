### Pass by Reference vs Value

All parameters (arguments) in the Python language are passed by reference. It means if you change what a parameter refers to within a function, the change also reflects back in the calling function. 

#### Example 3

```python

def changeme_1( mylist ):
   #This changes a passed list into this function
   mylist.append([1,2,3,4]);
   print "Values inside the function: ", mylist
   return

'''
    We maintain the reference of the passed object and 
    append values to the same object
'''

mylist_1 = [10,20,30];
changeme_1( mylist_1 );
print "Values outside the function: ", mylist_1
```

#### Example 4

```python
def changeme_2( mylist ):
   # This changes a passed list into this function
   mylist = [1,2,3,4]; # This would assign new reference in mylist
   print "Values inside the function: ", mylist
   return

'''
   The parameter mylist is local to the function changeme_2. 
   Changing mylist within the function does not affect mylist.
'''

mylist_2 = [10,20,30];
changeme_2( mylist_2 );
print "Values outside the function: ", mylist_2
```
